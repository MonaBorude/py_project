```python
# Replace closed woth Solved        and     pending with open
```

```python
# List Comprehension

# list_name = [var  if else   for i in list]
```

```python
df['New Status'] = ['Solved'   if i == 'Solved' or i == 'Closed'  else 'Open'   for i in df['Status'] ]
```

```python
df.head()
```

```python
# Remove Column Status   drop()

df1 = df.drop('Status', axis = 1)
```

```python
df1.head(1)
```

```python
# Data Analysis : 
```

```python
# 1. Display Complaint Status

df1['New Status'].value_counts()
```

```python
# plot bar plot.

plt.figure(figsize = (4,4))
df1['New Status'].value_counts().plot.bar()
plt.xlabel('Complaint Status')
plt.ylabel('Complaint Count')
plt.title('Customer Complaint Status')
```

```python
# 2.# plot " Daily Complaint chart "

df1['Date'].value_counts().plot()
```

```python
# 3.  City wise complaint [ top 10 ]

df1['City'].value_counts()[ : 10]
```

```python
df1['City'].value_counts()[ : 10].plot.bar()
```

```python
# 4. Display states with max complaints

df1['State'].value_counts()[ : 10]
```

```python
# 5. Display State wise Complaint status.

df1.groupby(['State'])['New Status'].value_counts().unstack()
```

```python
plt.figure(figsize = (15, 8))
df1.groupby(['State'])['New Status'].value_counts().unstack().plot.bar()
plt.show()
```
